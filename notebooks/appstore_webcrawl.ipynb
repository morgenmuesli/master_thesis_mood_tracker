```python
df = pd.read_excel("app_results_duplicates.xlsx")
```

```python
df.head(10)
```

```python
cleaned = df.drop_duplicates(subset=['appId'])
```

```python
cleaned["genre"].value_counts()
```

```python
genres = ["Lifestyle", "Medical", "Health & Fitness"]
```

```python
cleaned_genre = cleaned[cleaned["genre"].isin(genres)]
```

```python
cleaned_genre.count()
```

```python
from google_play_scraper import Sort, reviews_all
```

```python
cleaned_genre["number_of_reviews"] = cleaned_genre["appId"].apply(lambda x: len(reviews_all(x, sleep_milliseconds=0, lang='en', country='us', sort=Sort.MOST_RELEVANT)))
```

```python
cleaned_genre
```

```python
clean = cleaned_genre[cleaned_genre["number_of_reviews"] > 100]
```

```python
clean = clean[clean["free"] == 1]
```
