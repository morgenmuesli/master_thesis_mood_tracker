```python
import itertools
```

```python
fst_word = ["mood", "emotions", "feelings", "affective", "happines"]
snd_word = ["track", "journal", "record", "diary"]



tuples = itertools.product(fst_word, snd_word)


def scrape_all_searches(search_queries):
    df = pd.DataFrame
    for query in queries:
        result = search(query)
        if df.empty:
            df = pd.DataFrame.from_records(result)
        else:
            tmp = pd.DataFrame.from_records(result)
            df = pd.concat([df, tmp])
    return df
```

```python
queries = [f"{x[0]} {x[1]}" for x in tuples]
```

```python
data = scrape_all_searches(queries)
```

```python
data.to_excel("app_results_all.xlsx")
data.to_pickle("app_results_all.pkl")
```

```python
data_removed_duplicates = data.drop_duplicates(subset=['appId'])
```
